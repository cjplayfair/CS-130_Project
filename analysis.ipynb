```python
# Load data from drugs.csv into the program
def load_data():
    list_data = []
    for line in open('drugs.csv'):
        cleaned_line = line.strip()
        cleaned_list = [item.strip('"') for item in cleaned_line.split(',')]
        list_data.append(cleaned_list)
    list_data.pop(0)
    return list_data
data = load_data()

# Load New York data from drugs.csv into the program
def load_data_NY():
    list_data = []
    for line in open('drugs.csv'):
        cleaned_line = line.strip()
        cleaned_list = [item.strip('"') for item in cleaned_line.split(',')]
        if 'New York' in cleaned_list:
            list_data.append(cleaned_list)
    return list_data
data2 = load_data_NY()
```

```python
'''
How do substance abuse rates in New York compare to other states?
Create an averaged rate combining all substances for New York State. 
Assign a ranking list. 
Loop through every other state and if the state has a higher averaged rate, put it higher on the ranking list.
'''

def get_average_ny(data2):
    results_data_ny = {}
    for nested_list in data2:
        # Get year
        year = nested_list[1]
        # Note tobacco and cigarettes used past month rate as they do not have a past year rate
        # Get alcohol rates
        average_alcohol_12_17 = "Alcohol 12-17: " + nested_list[8]
        average_alcohol_18_25 = "Alcohol 18-25: " + nested_list[9]
        average_alcohol_26 = "Alcohol 26+: " + nested_list[10]
    
        # Get cigarette rates
        average_cigarette_12_17 = "Cigarettes 12-17: " + nested_list[20]
        average_cigarette_18_25 = "Cigarettes 18-25: " + nested_list[21]
        average_cigarette_26 = "Cigarettes 26+: " + nested_list[22]
            
        # Get cocaine rates
        average_cocaine_12_17 = "Cocaine 12-17: " + nested_list[26]
        average_cocaine_18_25 = "Cocaine 18-25: " + nested_list[27]
        average_cocaine_26 = "Cocaine 26+: " + nested_list[28]

        # Get marijuana rates
        average_marijuana_12_17 = "Marijuana 12-17: " + nested_list[32]
        average_marijuana_18_25 = "Marijuana 18-25: " + nested_list[33]
        average_marijuana_26 = "Marijuana 18-25: " + nested_list[34]
    
        # Get tobacco rates
        average_tobacco_12_17 = "Tobacco 12-17: " + nested_list[50]
        average_tobacco_18_25 = "Tobacco 18-25: " + nested_list[51]
        average_tobacco_26 = "Tobacco 26+: " + nested_list[52]
        results_value = [average_alcohol_12_17, average_alcohol_18_25, average_alcohol_26, average_cigarette_12_17, average_cigarette_18_25, average_cigarette_26, average_cocaine_12_17, average_cocaine_18_25, average_cocaine_26, average_marijuana_12_17, average_marijuana_18_25, average_marijuana_26, average_tobacco_12_17, average_tobacco_18_25, average_tobacco_26]
        results_data_ny[year] = results_value
    return results_data_ny
NY_averages = get_average_ny(data2)
#print(NY_averages)
```
